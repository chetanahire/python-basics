12.	Write a program to find the volume of the cylinder. Also find the cost when ,when the cost of 1 litre milk is 40Rs.

```python
# Volume = base area * hieght
# Volume =  πr2 × h
```

```python
import math

def volume_cylinder(h, r):
    vol_cm_cube =  math.pi * (r ** 2) * h
    return vol_cm_cube / 1000

def find_cost(vol_liters, price_1liter):
    return vol_liters * price_1liter

#driver code

h = 22
r = 25
vol = volume_cylinder(h, r)
print('The valoume of cylinder is: {} liter and cost of milk stored is : {}'.format(vol,  find_cost(vol, 40)))
```

16.	Write a program that will check whether the number is armstrong number or not.

```python
num = int(input("Enter the number"))
temp = num
add = 0
while temp > 0:
    rem = temp % 10
    add = add + rem ** 3
    temp = temp // 10

if num == add:
    print('Armstrong number')
else:
    print('Not Armstrong number')
```

19.	Write a menu driven program - 1.cm to ft  2.kl to miles  3.usd to inr  4.exit
-  1 cm = 1/30.48 feet
-  1 kilometer = 0.62137119 miles.
-  1 USD = 82.83 INR

```python

class MenuDriven:
    def __init__(self):
        self.__menu()
    
    def __menu(self):
        user_input = input("""
        Hello, How would you like to procced?
        
        1. Enter 1 to convert CM to Ft
        2. Enter 2 to convert KM to Miles
        3. Enter 3 to convert USD to INR
        4. Enter 4 to Exit
        
        """)
        if user_input == "1":
            self.convert_cm_to_ft()
        elif user_input == "2":
            self.convert_km_to_miles()
        elif user_input == "3":
            self.convert_usd_to_inr()
        elif user_input == "4":
            print('Bye')

    def convert_cm_to_ft(self):
        input_cm = float(input("Enter measurement in CM: \n"))
        print("{} CM  =  {} feet".format(input_cm, input_cm/30.48)) 
        self.__menu()

    def convert_km_to_miles(self):
        input_km = float(input("Enter Distance in KM : \n"))
        print("{} KM  =  {} Miles".format(input_km, input_km *  0.62137119))
        self.__menu()
    
    def convert_usd_to_inr(self):
        input_usd = float(input("Enter Amount in USD: \n"))
        print("{} USD  =  {} INR".format(input_usd, input_usd * 82.83))
        self.__menu()


    
# driver code
m = MenuDriven()
```
